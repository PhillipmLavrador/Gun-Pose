```python
import os
import csv
from ultralytics import YOLO

input_images_path = './input/'
annotations_csv_path = 'annotations.csv'

model = YOLO('./yolov8x-pose-p6.pt')
imgPathList = [f for f in os.listdir(input_images_path) if os.path.isfile(os.path.join(input_images_path, f))]
headerlist = ['Name','Dangerous','Keypoints', 'Confidence Levels']
nonDangerousPoses = 0
dangerousPoses = 0

with open(annotations_csv_path, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f ,delimiter =";")
    for img in imgPathList:
        if "Shotgun" in img or "Pistol" in img or "Rifle" in img:
            dangerousPoses += 1
            Dangerous = True
        else:
            nonDangerousPoses += 1
            Dangerous = False
        results = model(f"{input_images_path}{img}", stream=False, boxes=False, show=False)
        if(len(results[0].keypoints[0].xy[0]) < 2):
            continue
        else:
            coor = [[0]*2 for i in range(17)]
            conf = []
            for row in range(17):
                for col in range(2):
                    coor[row][col]=(results[0].keypoints[0].xyn[0][row][col].item())
            for i in range (17):
                conf.append(results[0].keypoints[0].conf[0][i].item())
            data = [img, Dangerous, coor, conf]
        writer.writerow(data)
        
print("Total Non Dangerous Poses Extracted" + str(nonDangerousPoses))
print("Total Dangerous Poses Extracted" + str(dangerousPoses))
```
